```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create


# Read in the data frame (df)

df = pd.read_csv(r'C:\Users\tripl\Documents\Courses\Alex The Analyst\Portfolio Project\Project 2\movies.csv')
```

## Data Cleaning

```python
df.head()
```

```python
# Look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# check for Negative Values on Budget Column
for val in df['budget']:
    if val < 0:
        print(val)
```

```python
# Transforming NA values to 0 in gross and budget columns

df['gross'] = df['gross'].fillna(0)

df['budget'] = df['budget'].fillna(0)

df 
```

```python
# Changing data type to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64') 
```

```python
# High correlation between budget and gross
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt. ylabel ('Budget for Film')

plt.show()
```

```python
# Tranforming text (string) data to numeric value

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# High correlation between budget and gross
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt. ylabel ('Budget for Film')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
